```python
from PIL import Image, ImageFilter
im = Image.open('pic.jpg')
w, h = im.size
print('Original image size: %sx%s' % (w, h))

# thumbnail函数接受一个元组作为参数，分别对应着缩略图的宽高，
# 在缩略时，函数会保持图片的宽高比例。如果输入的参数宽高和原图像宽高比不同，
# 则会依据最小对应边进行原比例缩放。
im.thumbnail((w//6, h//6))  # 生成缩略图

im_ = im.resize((100, 100))
print(im.size)
im.save('thumbnail.png', 'png')
```

```python
im
```

- <font size=4>ImageFilter.BLUR为模糊滤波，处理之后的图像会整体变得模糊。

```python
im2 = im.filter(ImageFilter.BLUR)  # 模糊效果
im2
```

- <font size=4>ImageFilter.CONTOUR为轮廓滤波，将图像中的轮廓信息全部提取出来。

```python
im3 = im.filter(ImageFilter.CONTOUR)  #  ImageFilter.CONTOUR为轮廓滤波，将图像中的轮廓信息全部提取出来。
im3
```

- <font size=4>ImageFilter.DETAIL为细节增强滤波，会使得图像中细节更加明显。

```python
im4 = im.filter(ImageFilter.DETAIL)
im4
```

- <font size=4>ImageFilter.EMBOSS为浮雕滤波，会使图像呈现出浮雕效果。

```python
im5 = im.filter(ImageFilter.EMBOSS)
im5
```

- <font size=4>ImageFilter.FIND_EDGES为寻找边缘信息的滤波，会找出图像中的边缘信息。

```python
im6 = im.filter(ImageFilter.FIND_EDGES)
im6
```

- <font size=4>ImageFilter.SMOOTH为平滑滤波，突出图像的宽大区域、低频成分、主干部分或抑制图像噪声和干扰高频成分，使图像亮度平缓渐变，减小突变梯度，改善图像质量。

```python
im7 = im.filter(ImageFilter.SMOOTH)
im7
```

- <font size=4>ImageFilter.SMOOTH_MORE为深度平滑滤波，会使得图像变得更加平滑。

```python
im8 = im.filter(ImageFilter.SMOOTH_MORE)
im8
```

- <font size=4>ImageFilter.SHARPEN为锐化滤波，补偿图像的轮廓，增强图像的边缘及灰度跳变的部分，使图像变得清晰。

```python
im9 = im.filter(ImageFilter.SHARPEN)
im9
```

- <font size=4>ImageFilter.EDGE_ENHANCE为边缘增强滤波，突出、加强和改善图像中不同灰度区域之间的边界和轮廓的图像增强方法。经处理使得边界和边缘在图像上表现为图像灰度的突变,用以提高人眼识别能力。

```python
im10 = im.filter(ImageFilter.EDGE_ENHANCE)
im10
```

- <font size=4>ImageFilter.EDGE_ENHANCE_MORE为深度边缘增强滤波，会使得图像中边缘部分更加明显。

```python
im11 = im.filter(ImageFilter.EDGE_ENHANCE_MORE)
im11
```
