```python
import tkinter as tk
from tkinter import filedialog
from tkinter import messagebox
from PIL import Image, ImageTk
window = tk.Tk()

'''框架'''

FrameA = tk.LabelFrame(window, text='FrameA\n你最喜欢中国四大美女中的哪一位？', padx=5, pady=5)  # padx pady设置框内边距
FrameA.grid(row=0, column=0, padx=10, pady=10)  # 框外边距
FrameB = tk.LabelFrame(window, text='FrameB\n登录界面', padx=5, pady=5)
FrameB.grid(row=0, column=1, padx=10, pady=10)
FrameC = tk.LabelFrame(window, text='FrameC\n加载图片', padx=5, pady=5)
FrameC.grid(row=0, column=2, padx=10, pady=10)
FrameD = tk.LabelFrame(window, text='FrameD\n', padx=5, pady=5)
#sticky 类似其他语言锚定功能，分东南西北四个方向，当全部锚定时控件或拉伸以填充整个网格
FrameD.grid(row=1, column=0, padx=10, pady=10, sticky='wesn')  
FrameE = tk.LabelFrame(window, text='FrameE\n', padx=5, pady=5)
FrameE.grid(row=1, column=1, padx=10, pady=10, sticky='wesn')

'''FrameA选择控件'''
girls =[('西施',1), ('王昭君',2), ('杨玉环',3), ('貂蝉', 4)]
v_a = tk.IntVar()
v_a.set(1)
for girl, num in girls:   
    tk.Radiobutton(FrameA,text = girl,value = num, variable = v_a).pack(anchor = 'w')

'''FrameB登录界面'''
tk.Label(FrameB, text='用户名').grid(row=0, column=0)
tk.Entry(FrameB).grid(row=0, column=1)
tk.Label(FrameB, text='密码').grid(row=1, column=0, sticky='w')
tk.Entry(FrameB, show='*').grid(row=1, column=1)
tk.Button(FrameB, text='登录').grid(row=2, columnspan=3)
img_b = Image.open('temp/3.jpg')
img_b.thumbnail((100,60))
tk_img_b = ImageTk.PhotoImage(img_b)
tk.Label(FrameB, image=tk_img_b).grid(row=0, column=2, rowspan=2, padx=5, pady=5)

'''FrameC加载图片'''
def choose():
    path_ = filedialog.askopenfilename(        # 可以不加参数
        title='选择文件', filetypes=
        [('JPEG', '*.jpg *.jpeg *.jpe *.jfif '),
        ('GIF', '*.gif'),
        ('PNG', '*.png'),
        ('ALL FILE', '*')
        ])    
    path_c.set(path_)
    img = Image.open(e_c.get())
    img.thumbnail((100, 100))
    tk_img = ImageTk.PhotoImage(img)
    l_c.config(image=tk_img)
    l_c.image = tk_img

FrameC_1 = tk.Frame(FrameC)
FrameC_1.grid(padx=10, pady=10,sticky='w')    
FrameC_2 = tk.Frame(FrameC)
FrameC_2.grid(padx=10, pady=10)

path_c = tk.StringVar()
b_c = tk.Button(FrameC_1, text='浏览', command=choose)
b_c.grid(row=0, column=0, sticky='w')
e_c = tk.Entry(FrameC_1, text=path_c, state='readonly')
e_c.grid(row=0, column=1, sticky='w')

l_c = tk.Label(FrameC_2)
l_c.pack()

'''FrameD按钮弹窗'''
var_d1 = tk.StringVar()
on_hit_d1 = False
on_change = False
def hit_me_d1():
    global on_hit_d1
    on_hit_d1 = not on_hit_d1
    if on_hit_d1 == True:
        var_d1.set('you hit me!')
    else:
        var_d1.set('')

def changelabel():
    global on_change
    on_change = not on_change
    if on_change == True:
        msg, color = 'goodbye world', 'blue'
    else:
        msg, color = 'hellow world', 'red'
    label_d2.config(text=msg, fg=color)
        
def callmsg():
    if messagebox.askokcancel(title='温馨提示', message='确定要退出吗？'):
        window.destroy()     
    
FrameD_1 = tk.Frame(FrameD)
FrameD_1.grid(sticky='ew')
FrameD_2 = tk.Frame(FrameD)
FrameD_2.grid(sticky='ew')
FrameD_3 = tk.Frame(FrameD)
FrameD_3.grid(sticky='ew')

button_d1 = tk.Button(FrameD_1, text='hit me', comman=hit_me_d1)
button_d1.pack(side='left', padx=5, pady=5)
label_d1 = tk.Label(FrameD_1, textvariable=var_d1, bg='white',width=20)
label_d1.pack(side='left', padx=5, pady=5)

button_d2 = tk.Button(FrameD_2, text='修改', command=changelabel)
button_d2.grid(row=0, column=0, padx=5, pady=5)
label_d2 = tk.Label(FrameD_2, text='helle world', bg='yellow', fg='red', width=20)
label_d2.grid(row=0, column=1, padx=5, pady=5)
    
button_d3 = tk.Button(FrameD_3, text='退出', command=callmsg)
button_d3.grid(row=1, column=0, padx=5, pady=5)    
    
'''FrameE inser'''
def insert_point():
    text_e.insert('insert', entry_e.get())

def insert_end():
    text_e.insert('end', entry_e.get())

entry_e = tk.Entry(FrameE)
entry_e.pack(fill='x')
tk.Button(FrameE, text='insert point',font=('隶书', 12), bg='yellow', fg='red', command=insert_point).pack(fill='x')
tk.Button(FrameE, text='insert end', command=insert_end).pack(fill='x')
text_e = tk.Text(FrameE, width=30, height=5)
text_e.pack(fill='x')





window.mainloop()
```
