## 方法 seek()

```python
import tkinter as tk
from PIL import Image, ImageTk
'''计算gif帧数'''
def frame_num(im):   
    try:
        n = -1
        while True:
            n = n + 1
            im.seek(n)          
    except:
        return n
'''逐帧显示'''
def run(n):
        global photo
        im = Image.open('test.gif')
        fn = frame_num(im)
        im.seek(n % fn)
        photo = ImageTk.PhotoImage(im)
        label.config(image=photo)
        root.after(500, run, n+1)
n = 0
root = tk.Tk()        
label = tk.Label(root)
label.pack()
run(n)
root.mainloop()
```

## 方法3

```python
from tkinter import *
import time
import os

def update(idx): # 定时器函数
    frame = frames[idx]
    idx += 1 # 下一帧的序号：在0,1,2,3,4,5之间循环(共6帧)
    label.configure(image=frame) # 显示当前帧的图片
    root.after(500, update, idx%numIdx) # 0.1秒(100毫秒)之后继续执行定时器函数(update)

root = Tk()
numIdx = 4 # gif的帧数
# 填充6帧内容到frames
frames = [PhotoImage(file=r"test.gif", format='gif -index %i' %(i)) for i in range(numIdx)]
label = Label(root)
label.pack()
root.after(0, update, 0) # 立即启动定时器函数(update)
root.mainloop()
```
